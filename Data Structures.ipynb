```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import NoSuchElementException
import time
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
data = pd.DataFrame()

# Start timer
start_time = time.time()

# Setup WebDriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

# Loop through pages 1 to 20
columns = ["Production Date", "Name", "Distance", "Type", "Fuel", "Transmission", "Color", "Interior color", "Exterior color", "Post Created", "Location", "Customs cleared", "Price"]
data = pd.DataFrame(columns=columns)

for page_number in range(1, 4):
    page_time = time.time()

    url = f'https://auto.am/search/passenger-cars?q={{%22category%22:%221%22,%22page%22:%22{page_number}%22,%22sort%22:%22latest%22,%22layout%22:%22list%22,%22user%22:{{%22dealer%22:%220%22,%22official%22:%220%22,%22id%22:%22%22}},%22year%22:{{%22gt%22:%221911%22,%22lt%22:%222026%22}},%22usdprice%22:{{%22gt%22:%220%22,%22lt%22:%22100000000%22}},%22mileage%22:{{%22gt%22:%2210%22,%22lt%22:%221000000%22}}}}'

    driver.get(url)
    time.sleep(5)  # Wait for the page to load

    # Find all car listings on the page
    car_links = driver.find_elements(By.CLASS_NAME, 'card')

    for counter, car in enumerate(car_links, start=1):
        car_data = {}

        try:
            title = car.find_elements(By.CLASS_NAME, "card-title")
            if title:
                title_list = title[0].text
                title_elements = title_list.split(" ")
            else:
                title_elements  = []
        except NoSuchElementException:
            title_elements = []

        car_data["Production Date"] = title_elements[0] if len(title_elements) > 0 else "Unknown"
        car_data["Name"] = title_elements[1] if len(title_elements) > 1 else "Unknown"

        try:
            desc_elements = car.find_elements(By.CLASS_NAME, "card-desc")
            if desc_elements:
                desc = desc_elements[0].text
                desc_list = desc.split(", ")
            else:
                desc_list = []
        except NoSuchElementException:
            desc_list = []
        
        car_data["Distance"] = desc_list[0] if len(desc_list) > 0 else "Unknown"
        car_data["Type"] = desc_list[1] if len(desc_list) > 1 else "Unknown"
        car_data["Fuel"] = desc_list[2] if len(desc_list) > 2 else "Unknown"
        car_data["Transmission"] = desc_list[3] if len(desc_list) > 3 else "Unknown"
        car_data["Color"] = desc_list[4] if len(desc_list) > 4 else "Unknown"
        
        # Handle optional color values
        if len(desc_list) > 5:
            car_data["Interior color"] = desc_list[5]
        else:
            car_data["Exterior color"] = "Black"
                
        try:
            location = car.find_elements(By.CLASS_NAME, "card-loc")
            if location:
                loc = location[0].text
                loc_list = loc.split(", ")
            else:
                loc_list = []
        except NoSuchElementException:
            loc_list = []

        if loc_list:
            post_loc = loc_list[0]
            post_loc_list = post_loc.split(" ")
        
            car_data["Post Created"] = post_loc_list[0] if len(post_loc_list) > 0 else "Unknown"
        
            if len(post_loc_list) > 1:
                car_data["Location"] = post_loc_list[1] 
        
        # Check customs clearance info safely
        if len(loc_list) > 1:
            car_data["Customs cleared"] = 1 if loc_list[1] == "Մաքսազերծված է" else 0
        else:
            car_data["Customs cleared"] = "Unknown"

        

        try:
            car_data['Price'] = car.find_element(By.CLASS_NAME, "price").text
        except NoSuchElementException:
            car_data['Price'] = "Null"

            # Append new data to the DataFrame
        data = pd.concat([data, pd.DataFrame([car_data])], ignore_index=True, sort=False)

        print(f"Processed {counter}-th car on page {page_number}")

    page_end_time = time.time()
    print(f'Time for page {page_number}: {page_end_time - page_time} seconds')

print("Data successfully written to scraped_data.csv")

end_time = time.time()
print(f'Total execution time: {end_time - start_time} seconds')

data.to_csv("cars_data.csv", index=False, encoding="utf-8")

# Close the WebDriver
driver.quit()
```
